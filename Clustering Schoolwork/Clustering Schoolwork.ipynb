```python
#确定邻接矩阵（带权值）
#首先定义一个列表
# L = []
for i in range(15):
    for j in range(15):
        #将每两个点之间的欧氏距离存储在列表内
        d = np.sqrt(np.sum(np.square(x[i]-x[j])))
        #print(d)
        L.append(d)#长度为22500 类型为list
#将列表转换成数组便于运算
L_A = np.array(L)#长度是22500
```

```python
G = nx.Graph()
L = []
for i in range(len(L_A)):
    for j in range(len(L_A)):
        G.add_edge(i,j)
nx.draw(G)
plt.show()
```
